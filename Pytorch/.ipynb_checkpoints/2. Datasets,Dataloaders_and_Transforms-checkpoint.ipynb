## Datasets and Dataloaders
Dataset and Dataloader abstraction in pytorch facilitates better readebility and modularity

- **Dataset** - stores the samples and their corresponding labels
- **Dataloader** - wraps an iterable around the Dataset to enable easy access to the samples.

Advantages of using Dataset and Dataloaders:
- keeps your data manageable and helps to simplify your machine learning pipeline
- No longer we need to worry about shuffling and slicing the data inorder retrieve batches.
- We need not load the whole dataset into memory.

## Transforms

We can define the preprocessing required for the data, before dataloader pushes it to model, using transforms parameters.

- `transform` - preprocessing to be applied on features
- `target_transform` - preprocessing to be applied on  labels

They accepts predefined transformation from ` torchvision.transforms` as well user defined transformations defined using `Lambda`

```python
import torch
from torchvision import datasets
from torchvision.transforms import ToTensor, Lambda

ds = datasets.FashionMNIST(
    root="data",
    train=True,
    download=True,
    transform=ToTensor(),
    target_transform=Lambda(lambda y: torch.zeros(10, dtype=torch.float).scatter_(0, torch.tensor(y), value=1))
)
```

> **`ToTensor()`** - *italicized text*
converts a PIL image or NumPy ndarray into a FloatTensor. and scales the image’s pixel intensity values in the range [0., 1.]

```python
Lambda(lambda y: torch.zeros(10, dtype=torch.float).scatter_(dim=0, index=torch.tensor(y), value=1))
```

> This function to turn the integer into a one-hot encoded tensor.
