## Activation functions

Activation functions are mathematical functions used in artificial neural networks to determine the output of a neuron, based on the inputs it receives. It behaves similar to excitation of human neurons. 
They introduce non-linearity into the output of a neuron, allowing neural networks to model complex relationships between inputs and outputs.

```python
import matplotlib.pyplot as plt
import numpy as np
```

### 1. Binary step function / Hard Threshold Activation
$$f(x) = \begin{cases}
  1 ,\ for \  x \geq \theta \\    
  0 ,\ for \  x <\theta    
\end{cases}$$

```python
def hardthreshold(x):
    if x>=0:
        return 1
    else:
        return 0

x=np.linspace(-5,5)
y=list(map(hardthreshold, x)) 

plt.title("Hard Threshold Activation")
plt.xlabel("Input")
plt.ylabel("Activation")
plt.plot(x, y)
plt.show()
```

Limitations:
- Not differentiable and thus gradients in the input cannot be determined

### 2. Sigmoid Activation
$$ f(x) = \frac{1}{1+e^{-x}}$$

```python
def sigmoid(x):
    return 1/(1 + np.exp(-x))

x=np.linspace(-5,5)
y=list(map(sigmoid, x)) 

plt.title("Sigmoid Activation")
plt.xlabel("Input")
plt.ylabel("Activation")
plt.plot(x, y)
plt.show()
```

Limitations:
- Output of all the neurons will be of the same sign.
- Gradient becomes negligible for very high value of |x|, and leads to vanishing gradient problem in neural networks.

### 3. Tanh Activation
$$ f(x) = \frac{e^{x}-e^{-x}}{e^{x}+e^{-x}}$$

  


```python
def tanh(x):
    return ( np.exp(x) - np.exp(-x) )/(np.exp(x) + np.exp(-x))

x=np.linspace(-5,5)
y=list(map(tanh, x)) 

plt.title("Tanh Activation")
plt.xlabel("Input")
plt.ylabel("Activation")
plt.plot(x, y)
plt.show()
```

- Now, the Output of the neurons can be both postive and negative.

*Limitations:*
- Has same problem as sigmoid

### 4. ReLU Activation
$$ f(x) = max(0,x)$$


```python
def relu(x):
    return max(0,x)

x=np.linspace(-5,5)
y=list(map(relu, x)) 

plt.title("ReLU Activation")
plt.xlabel("Input")
plt.ylabel("Activation")
plt.plot(x, y)
plt.show()
```

- No vanishing gradient issue.
- Computationally faster than sigmoid and tanh

*Limitations:*
- Zero gradient for negative input, prevents updates of weights of some neuron( ,called dead neurons).

### 5. Leaky ReLU Activation
$$ f(x) = max(0.1x, x)$$


```python
def leakyRelu(x):
    return max(0.1*x,x)

x=np.linspace(-5,5)
y=list(map(leakyRelu, x)) 

plt.title("Leaky ReLU Activation")
plt.xlabel("Input")
plt.ylabel("Activation")
plt.plot(x, y)
plt.show()
```

- We get gradient for negative input as well, and thus allow backporpagation.

*Limitations:*
- Gradients of negative inputs is very small and thus learning is slow.

### 6. ELU Activation
$$f(x) = \begin{cases}
  x ,\ \ \ \ \ \  \ \ \ \ \ \ \ \  for \  x \geq 0 \\    
  \alpha(e^x - 1) ,\ for \  x <0    
\end{cases}$$

```python
def Elu(x):
    if x>=0:
        return x
    else:
        alpha =1
        return alpha*(np.exp(x) -1)

x=np.linspace(-5,5)
y=list(map(Elu, x)) 

plt.title("ELU Activation")
plt.xlabel("Input")
plt.ylabel("Activation")
plt.plot(x, y)
plt.show()
```
